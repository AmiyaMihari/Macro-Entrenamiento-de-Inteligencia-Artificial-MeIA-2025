```python
#Importar librerÃ­as
from transformers import pipeline
import gradio as gr
```

```python
# Frases de ejemplo para cada intenciÃ³n
intents = {
    "saludo": ["hola", "buenos dÃ­as", "quÃ© tal", "hey"],
    "horario": ["Â¿cuÃ¡l es el horario?", "Â¿a quÃ© hora abren?", "dime tu horario de atenciÃ³n", "puedo llamarte en la noche?"]
}
```

```python
# Etiquetas y respuestas asociadas
responses = {
    "saludo": "Â¡Hola! Â¿En quÃ© puedo ayudarte?",
    "horario": "Nuestro horario de atenciÃ³n es de lunes a viernes, de 9 a.m. a 6 p.m.",
    "desconocido": "Lo siento, no entendÃ­ tu mensaje. Â¿PodrÃ­as reformularlo?"
}
```

```python
# Clasificador de intenciones usando zero-shot
classifier = pipeline("zero-shot-classification", model="joeddav/xlm-roberta-large-xnli")
```

```python
def chatbot(input_text):
    labels = list(intents.keys())
    result = classifier(input_text, labels)
    top_label = result["labels"][0]
    score = result["scores"][0]
    if score < 0.6:
        return responses["desconocido"]
    return responses.get(top_label, responses["desconocido"])
```

```python
# ðŸ’¬ Interfaz con Gradio
iface = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(lines=2, placeholder="Escribe tu mensaje..."),
    outputs="text",
    title="ðŸ¤– Chatbot en EspaÃ±ol",
    description="Chatbot bÃ¡sico que reconoce intenciones (saludo, horario)."
)
```

```python
iface.launch()
```
