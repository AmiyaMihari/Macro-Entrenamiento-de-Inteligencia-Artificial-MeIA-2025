```python
#Importar librerías
from transformers import pipeline
import gradio as gr
```

```python
# Frases de ejemplo para cada intención
intents = {
    "saludo": ["hola", "buenos días", "qué tal", "hey"],
    "horario": ["¿cuál es el horario?", "¿a qué hora abren?", "dime tu horario de atención", "puedo llamarte en la noche?"]
}
```

```python
# Etiquetas y respuestas asociadas
responses = {
    "saludo": "¡Hola! ¿En qué puedo ayudarte?",
    "horario": "Nuestro horario de atención es de lunes a viernes, de 9 a.m. a 6 p.m.",
    "desconocido": "Lo siento, no entendí tu mensaje. ¿Podrías reformularlo?"
}
```

```python
# Clasificador de intenciones usando zero-shot
classifier = pipeline("zero-shot-classification", model="joeddav/xlm-roberta-large-xnli")
```

```python
def chatbot(input_text):
    labels = list(intents.keys())
    result = classifier(input_text, labels)
    top_label = result["labels"][0]
    score = result["scores"][0]
    if score < 0.6:
        return responses["desconocido"]
    return responses.get(top_label, responses["desconocido"])
```

```python
# 💬 Interfaz con Gradio
iface = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(lines=2, placeholder="Escribe tu mensaje..."),
    outputs="text",
    title="🤖 Chatbot en Español",
    description="Chatbot básico que reconoce intenciones (saludo, horario)."
)
```

```python
iface.launch()
```
