```python
from transformers import pipeline
import pandas as pd
```

```python
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
contexto = """El Teorema de Pitágoras es un principio fundamental en geometría que establece que en un triángulo
rectángulo, el cuadrado de la hipotenusa es igual a la suma de los cuadrados de los otros dos lados.
Esta relación se expresa con la fórmula: a² + b² = c²."""
```

```python
pregunta = "¿Cuál es la relación entre los lados de un triángulo rectángulo según el Teorema de Pitágoras?"
```

```python
resultados = qa_pipeline(question=pregunta, context=contexto, topk=3)
for r in resultados:
    print(f"{r['answer']} ({r['score']:.2f})")
```

```python
modelos = {
    "distilbert": "distilbert-base-cased-distilled-squad",
    "bert": "bert-large-uncased-whole-word-masking-finetuned-squad",
    "roberta": "deepset/roberta-base-squad2",
    "electra": "deepset/electra-base-squad2"
}

resultados = []
for nombre, modelo in modelos.items():
    qa = pipeline("question-answering", model=modelo, tokenizer=modelo)
    r = qa(question=pregunta, context=contexto)
    resultados.append({"Modelo": nombre, "Respuesta": r["answer"], "Score": round(r["score"], 4)})
```

```python
print(pd.DataFrame(resultados))
```
