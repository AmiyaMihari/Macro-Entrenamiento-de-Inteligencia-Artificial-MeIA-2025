```python
import gradio as gr
from transformers import pipeline
```

```python
# Inicializar modelo QA
#qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
# Documentos predefinidos
documentos = {
    "Alan Turing": """Alan Turing fue un matemático británico que trabajó en Bletchley Park durante la Segunda Guerra Mundial.
                      Ayudó a descifrar la máquina Enigma. También propuso una prueba para evaluar si una máquina puede imitar
                      la inteligencia humana, conocida como prueba de Turing.""",
    "Albert Einstein": """Albert Einstein fue un físico teórico alemán conocido por desarrollar la teoría de la relatividad.
                          Su ecuación E=mc² es una de las más famosas de la física. Recibió el Premio Nobel en 1921.""",
    "Isaac Newton": """Isaac Newton formuló las leyes del movimiento y la ley de la gravitación universal.
                       Escribió 'Philosophiæ Naturalis Principia Mathematica', una obra clave en la historia de la ciencia."""
}
```

```python
# Función para responder preguntas
def responder(pregunta, documento):
    contexto = documentos[documento]
    if not pregunta.strip():
        return "Por favor escribe una pregunta."
    resultado = qa_pipeline(question=pregunta, context=contexto)
    return resultado["answer"]
```

```python
# Interfaz Gradio
iface = gr.Interface(
    fn=responder,
    inputs=[
        gr.Textbox(label="Escribe tu pregunta"),
        gr.Dropdown(choices=list(documentos.keys()), label="Selecciona un documento")
    ],
    outputs="text",
    title="Sistema de Preguntas y Respuestas",
    description="Haz una pregunta sobre uno de los documentos disponibles. El modelo responderá usando sólo el contenido seleccionado."
)
```

```python
iface.launch()
```
