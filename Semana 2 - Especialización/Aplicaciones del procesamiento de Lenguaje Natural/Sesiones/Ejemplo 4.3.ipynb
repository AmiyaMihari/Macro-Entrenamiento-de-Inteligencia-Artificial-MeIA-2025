```python
from sentence_transformers import SentenceTransformer, util
from transformers import pipeline
import pandas as pd
```

```python
documentos = {
    "turing": """Alan Turing fue un pionero en la ciencia de la computación. Durante la Segunda Guerra Mundial,
                 descifró el código Enigma. También propuso la prueba de Turing, considerada una de las primeras ideas
                 formales sobre inteligencia artificial.""",
    "mccarthy": """John McCarthy fue un científico estadounidense que acuñó el término inteligencia artificial (IA) en 1956.
                   Desarrolló el lenguaje de programación Lisp y propuso la idea de máquinas que razonan como humanos.""",
    "einstein": """Albert Einstein es conocido por la teoría de la relatividad y la fórmula E=mc². Su trabajo revolucionó
                   la física moderna, pero no estuvo directamente relacionado con la computación ni la inteligencia artificial.""",
    "vonneumann": """John von Neumann contribuyó a la creación de la arquitectura de computadores moderna.
                     Su diseño de arquitectura almacenada se usa en la mayoría de computadoras actuales y fue clave
                     para el desarrollo de la informática."""
}
```

```python
pregunta = "¿Qué científicos hicieron contribuciones fundamentales a la inteligencia artificial o la computación?"
```

```python
#Embeddings semánticos
model_embed = SentenceTransformer('all-MiniLM-L6-v2')
doc_nombres = list(documentos.keys())
doc_textos = list(documentos.values())
doc_embs = model_embed.encode(doc_textos, convert_to_tensor=True)
query_emb = model_embed.encode(pregunta, convert_to_tensor=True)
scores = util.cos_sim(query_emb, doc_embs)[0]
```

```python
#Selección de los top 2 documentos
top_n = 2
indices_top = scores.topk(top_n).indices.tolist()
docs_seleccionados = [doc_nombres[i] for i in indices_top]
textos_contexto = [doc_textos[i] for i in indices_top]
contexto_completo = "\n".join(textos_contexto)
```

```python
gen_pipeline = pipeline("text2text-generation", model="google/flan-t5-base")
```

```python
#Modelo generativo (T5)
prompt = f"Contexto:\n{contexto_completo}\n\nPregunta: {pregunta}\n\nRespuesta:"
respuesta_generada = gen_pipeline(prompt, max_length=128, do_sample=False)[0]['generated_text']
```

```python
#Tabla de resultados
datos = {
    "Documentos seleccionados": ", ".join(docs_seleccionados),
    "Similitud promedio": round(scores[indices_top].mean().item(), 4),
    "Respuesta generada": respuesta_generada
}
```

```python
df_resultado = pd.DataFrame([datos])
```
