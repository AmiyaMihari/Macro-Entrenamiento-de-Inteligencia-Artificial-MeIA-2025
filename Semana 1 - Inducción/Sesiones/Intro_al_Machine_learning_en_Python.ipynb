# Regresión lineal (Simple)

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/350px-Linear_regression.svg.png">

* La regresión lineal es una técnica de modelado estadístico que se emplea para describir una variable de respuesta continua como una función de una o varias variables predictoras.

* Su formula es:

$$ Y = \beta_{0} + \sum \beta_{i}X_{i} + \epsilon_{i}$$

* En este caso podriamos decir que:
* PrecioCasa = $10 + ($50*room_num) + cantidaQueFaltoParaElPrecioReal

```python
#NUEVAS LIBRERIAS A UTILZIAR
import statsmodels.api as sm
```

```python
# AHORA VAMOS A SEPARAR EL CONJUNTO DE DATOS ENTRE CARACTERISTICAS Y TARGETS
X = sm.add_constant(df['room_num'])
```

# Ordinary Least Squares
regression-error.png

$$RSS = (y_{1}-\hat{\beta_{0}}-\hat{\beta_{1}}x_{1})^{2}+(y_{2}-\hat{\beta_{0}}-\hat{\beta_{1}}x_{2})^{2}+...+(y_{n}-\hat{\beta_{0}}-\hat{\beta_{1}}x_{n})^{2}$$

$$ \hat{\beta_{1}}=\frac{\sum^{n}_{i=1}(x_{i}-\bar{x})(y_{i}-\bar{y})}{\sum^{n}_{i=2}(x_{i}-\bar{x})^{2}}$$

$$\hat{\beta_0}=\bar{y}-\hat{\beta_1}\bar{x}$$

```python
# Ahora crearemos un modelo para predecir el precio, en base al numero de cuartos
# Deduciremos los valores de Beta0 y Beta_i en base a los ejemplos que tenemos en el dataset
# Dichos valores se obtienen mediante despejes. Como son muchos se utiliza el método de minimos cudrados
lm = sm.OLS(df['price'],X).fit()
```

```python
#QUE SALIO?
lm.summary()
```

## ¿Qué obtuvimos como resultado?
* Al construir el modelo (la regresión lineal simple), La salida nos muestra los valores que se determinaron para Beta_0 y Beta_1.
* Aparte de los valores es también necesario saber si existe una relación entre $$Y=\beta_{0}+\beta_{1}X + \epsilon$$
* A ojo podemos ver que el valor de B_1 (9.099) es diferente de 0, por lo cual esa relación existe. Aun así, ¿el valor es significante estadísticamente hablando?
* El software también realiza las pruebas estadísticas. El valor P reportado es menor a 0.05, por lo cual con un 95% de confianza sabemos que existe una relación entre X y Y.

# Podemos utilizar más variables para hacer la predicción?

## Si, utilizando una regresión lineal múltiple

* En ves de $$Y=\beta_{0}+\beta_{1}X + \epsilon$$

* Es $$ Y = \beta_{0} + \sum \beta_{i}X_{i} + \epsilon_{i}$$

```python
# Con código es algo así
X_multi = df.drop('price', axis=1) #nos quedamos con todas las variables, menos
                                   #precio
X_multi.head() #como se ve eso?
```

```python
y_multi = df['price'] #ahora creamos la variable a predecir
y_multi.head()
```

```python
#ahora, en este caso para construir el modelo es necesario
#crear una nueva columna para el Beta_0 llamada const (constante)
#(Este caso es para statsmodels, no para sklearn)
X_multi_cons = sm.add_constant(X_multi)
X_multi_cons.head()
```

```python
#Procedemos a crear el modelo con todas las variables que quedaron
lm_multi = sm.OLS(y_multi,X_multi_cons).fit()
lm_multi.summary()
```

* Como podemos ver, el sumario muestra los beta_1 de todas las variables
* Podemos ver que las variables crime_rate y age casi no aportan nada.
* Según el reporte del modelo, solo hay relación significante entre X y Y son:
      * air_qual
      * room_num
      * teachers
      * poor_prop
      * avg_dist


```python
#Ahora con sklearn
lm3 = LinearRegression()
lm3.fit(X_multi,y_multi)
```

```python
#Cuales fueron los valores de los coeficientes?
print(lm3.intercept_,lm3.coef_)
```

# ¿Como podemos ver que tan bien predice el modelo?
## Podemos crear una partición del conjunto de datos.
*	Una parte para deducir los coeficientes (Entrenamiento)
*	Otra parte para ver que tan bien predice (Prueba)
*	Que tanto es para cada uno? Generalmente es 70-30 u 80-20. Es cuestión de preferencias y disponibilidad de datos.


```python
# Llamaremos las librerias necesarias para hacer el proceso
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_multi,y_multi, test_size=0.2, random_state=0)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
#Creamos el nuevo objeto de regresion lineal múltiple
lm_a = LinearRegression()
lm_a.fit(X_train,y_train) # Le pasamos los datos con los que va a construir el modelo
                          # las variables y los precios. Asi puede ajustar hasta que
                          # las predicciónes sean cercanos a los precios y el error
                          # sea el mínimo.
```

```python
# ahora vamos a hacer la predicción del modelo construido
# primero en el conjunto de entrenamiento
# luego en el de prueba
y_train_a = lm_a.predict(X_train)
y_test_a = lm_a.predict(X_test)
```

## que métrica se utilizará?
* $$R^{2}= 1- \frac{SS_{res}}{SS_{tot}}$$
* SS_res = Suma de los cuadrados de los errores residuales
* SS_tot = Suma total de los errores
* El resultado obtenido reflejara el porcentaje en el que el modelo puede explicar la variabilidad de la salida (el precio)
* Si la salida es 0.68 quiere decir que el 68% es explicado por el modelo, mientras que el 32% restante no lo puede explicar
* Entre más alto el valor mejor. Si es de 80% es muy bueno, de 50% es mejor lanzar una moneda para adivinar (no es bueno), si es de 100% es malo, por que no hay modelo perfecto, puede ser error al programar, sobreajuste o trampa intencional

```python
# importando la libreria para calcular la métrica
from sklearn.metrics import r2_score
```

```python
# ahora necesitamos medir que tan bien predice
# primero en training
# luego en test
rtrain = r2_score(y_train,y_train_a)
rtest  = r2_score(y_test,y_test_a)
print(f"\n El r en training es:{rtrain}\n El r en test es:{rtest}\n")
```

## Por que sale tan bien en training y tan mal en test?
* El modelo se construye con los datos de entrenamiento y los coeficientes se obtienen para disminuir el error en esa muestra.
* Los datos en test son desconocidos, por lo cual es común que exista un error más elevado que en training.
* Lo ideal es incrementar la cantidad de datos de entrenamiento para mejorar el modelo, sin embargo, no siempre es posible pues no hay datos suficientes.
